```python
def climbingLeaderboard(scores, alice):
    rank = list()
    for i in range(len(alice)):
        
        leader = list()
        a=1
        arr = scores.copy()
        arr.append(alice[i])
        arr = sorted(arr,reverse=True)
        for j in range(len(arr)):
            if j == 0:
                leader.append(a)
                a=a+1
            elif arr[j] == arr[j-1]:
                leader.append(leader[j-1])
            elif arr[j] != arr[j-1]:
                leader.append(a)
                a=a+1
                
            
        for k in range(len(arr)):
            if arr[k]==alice[i]:
                rank.append(leader[k])
                break
        
    return rank
scores = [100,100,50,40,40,20,10]
alice = [5,25,50,120]
rank = climbingLeaderboard(scores,alice)
print(rank)
```

```python
# Importing Various Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from statistics import *
```

# Problem 4.1

```python
# Lets import the stock of Ashoka
df = pd.read_csv("C:/Users/Tushar Goel/Desktop/Intern related dataset/Small_Cap/ASHOKA.csv")
```

```python
df.head()
```

```python
# Bollinger Columns 
rolling_std = df['Close Price'].rolling(7).std()
df['Rolling Mean'] = df['Close Price'].rolling(7).mean()
df['Upper Band'] = df['Close Price'].rolling(7).mean() + rolling_std*2
df['Lower Band'] = df['Close Price'].rolling(7).mean() - rolling_std*2
```

```python
df
```

### Lets treat the NaN values by filling them by their mean

```python
df['Rolling Mean'].fillna(df['Rolling Mean'].mean(),inplace=True)
df['Upper Band'].fillna(df['Upper Band'].mean(),inplace=True)
df['Lower Band'].fillna(df['Lower Band'].mean(),inplace=True)
```

```python
df
```

### Lets Create a Call column

```python
df['Call'] = np.zeros(494)
for i in range(df.shape[0]):
    if df['Close Price'][i]<df['Lower Band'][i]:
        df['Call'][i] = 'Buy'
    elif df['Close Price'][i]>df['Lower Band'][i] and df['Close Price'][i]<df['Rolling Mean'][i]:
        df['Call'][i] = 'Hold Buy/Liquidate Short'
    elif df['Close Price'][i]>df['Rolling Mean'][i] and df['Close Price'][i]<df['Upper Band'][i]:
        df['Call'][i] = 'Hold Short/Liquidate Buy'
    elif df['Close Price'][i]>df['Upper Band'][i]:
        df['Call'][i] = 'Short'
```

```python
df
```

```python
df['Call'].value_counts()
```

# Creating the Model

```python
x = df[['Close Price','Upper Band','Rolling Mean','Lower Band']]
x.head()
```

```python
y = df['Call']
y.head()
```

 ## Preprocessing of the Categorical data

```python
le = LabelEncoder()
y = le.fit_transform(y)
y = y.reshape(-1,1)
```

```python
#Applying One Hot Encoding
ohc = OneHotEncoder(categorical_features=[0])
y = ohc.fit_transform(y).toarray()
```

```python
y.shape
```

## Splitting of Data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=0)
```

# Logistic Regression using Hyperparameter Tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV # HyperParameter Tuning
```

```python
from scipy.stats import uniform
```

```python
lr = LogisticRegression()

penalty = ['l1','l2']
C = uniform(loc = 0,scale=5)
tol = [0.0001,0.001,0.01,0.1,1]

hyperparameters = dict(penalty=penalty,C=C,tol=tol)

lf = RandomizedSearchCV(lr,hyperparameters,n_iter = 200 , n_jobs=-1,cv=5,verbose=1)
```

```python
logistic_regression  = lf.fit(x_train,y_train)
```

```python
y_predict = logistic_regression.predict(x_test)
print(accuracy_score(y_test,y_predict))
```

## Logistic regression Model has 92 % accuracy

# Support Vector Machine using HyperParameter Tuning

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
kernel = ['rbf','poly','linear']
degree = [1,2,3,4,5,6]
C = uniform(loc=0,scale=5)
gamma = ['auto','scale']
tol = [0.0001,0.001,0.01,0.1,1]
hyperparameters = dict(kernel=kernel,degree=degree,C=C,tol=tol,gamma=gamma)
lg = RandomizedSearchCV(svc,hyperparameters)
```

```python
support_vector_classifier = lg.fit(x_train,y_train)
```

```python
y_predict_svc = support_vector_classifier.predict(x_test)
print(accuracy_score(y_test,y_predict_svc))
```

## Support Vector Machine has an accuracy of 96%

```python
# Asian Paint Stock
stock = pd.read_csv("C:/Users/Tushar Goel/Desktop/Intern related dataset/Large_Cap/ASIANPAINT.csv")
```

```python
stock.head()
```

## Formation of 4 Columns

```python
stock['Day_Perc_Change'] = stock['Close Price'].pct_change()
```

```python
stock.drop(0,inplace=True)
```

```python
stock['Rolling Mean']=stock['Day_Perc_Change'].rolling(5).mean()
stock['Rolling Std']=stock['Day_Perc_Change'].rolling(5).std()
```

```python
# Filling up NaN Values
stock['Rolling Mean'].fillna(stock['Rolling Mean'].mean(),inplace=True)
# Filling up NaN Values
stock['Rolling Std'].fillna(stock['Rolling Std'].mean(),inplace=True)
```

```python
stock['Open_Close_change'] = np.zeros(stock.shape[0])
stock['Low_high_change'] = np.zeros(stock.shape[0])
```

```python
for i in range(1,stock.shape[0]):
    stock['Open_Close_change'][i] = (stock['Close Price'][i]-stock['Open Price'][i])*100/(stock['Open Price'][i])
    stock['Low_high_change'][i] = (stock['High Price'][i]-stock['Low Price'][i])*100/(stock['Low Price'][i])
```

```python
stock.head()
```

## Formation of Target Function

```python
stock['Action'] = np.zeros(stock.shape[0])
```

```python
for i in range(1,stock.shape[0]):
    if stock['Close Price'][i+1]>stock['Close Price'][i]:
        stock['Action'][i] = int(1)
    else:
        stock['Action'][i] = int(-1)
```

```python
stock.head()
```

```python
x_stock = stock[['Open_Close_change','Low_high_change','Rolling Mean','Rolling Std']]
```

```python
y_stock = stock['Action']
```

```python
# Label Encoder
```

```python
les = LabelEncoder()
y_stock = les.fit_transform(y_stock)
```

```python
y_stock
```

```python
x_train_stock,x_test_stock,y_train_stock,y_test_stock = train_test_split(x_stock,y_stock,test_size=0.1,random_state=0)
```

## Random Forest Classifier Model

```python
model = RandomForestClassifier(criterion='gini')
```

```python
random_class = {'n_estimators':range(100,1000,10),
                'min_samples_split' : range(2,10)}

lh = RandomizedSearchCV(model,random_class).fit(x_train_stock,y_train_stock)
y_predict_model = lh.predict(x_test_stock)
```

```python
print(accuracy_score(y_test_stock,y_predict_model))
```

## Accuracy of the model is 56%

## as new is highly coorelated with OHLC , it is highly possible that it is linear related to them.
## Lets check it out by training our model


```python
y_1 = df['Pred'].loc[:410]
x_2 = df.iloc[:,[1,2,3,4]]
x_1 = x_2.loc[:410]
y_2 = df['new']
```

```python
y_1.isnull().value_counts()
```

## Now we are going to train our model

### First we going to determine which column is polynomial related or linear related to the inputs by comparing the accuracies (r2_score)
### we are goin to prepare our model for 'New' column

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# spliting the data
x_2_train,x_2_test,y_2_train,y_2_test = train_test_split(x_2,y_2,test_size=0.1,random_state=0)
```

```python
lr = LinearRegression()
```

```python
linear_model = lr.fit(x_2_train,y_2_train)
```

```python
y_2_predict = linear_model.predict(x_2_test)
```

```python
print(r2_score(y_2_test,y_2_predict))
```

## It is 99.99 % accurate . Hence "New" is Linear related
## Then Pred is Polynomially related

## Let's Train our model for Pred

```python
poly = PolynomialFeatures(3)

x_1 = poly.fit_transform(x_1)
```

```python
x_1_train,x_1_test,y_1_train,y_1_test = train_test_split(x_1,y_1,test_size=0.1,random_state=0)
```

```python
lr_1 = LinearRegression()
linear_model_1 = lr_1.fit(x_1_train,y_1_train)
```

```python
y_1_predict = linear_model_1.predict(x_1_test)
```

```python
print(r2_score(y_1_test,y_1_predict))
```

```python
linear_model_1.coef_
```

## This Polynomial model is 99.99% accurate (Pred Column)
## Let's fill our columns fields by this model

```python
value = df.iloc[:,[1,2,3,4]].loc[411:]
```

```python
value = poly.fit_transform(value)
```

```python
a = linear_model_1.predict(value)
```

```python
a
```

```python
df['Pred'].loc[411:] = a
```

```python
df['Pred']
```

# Beta Value for Monthly Return

## Monthly Close price for Nifty and Stock

```python
stock.reset_index(inplace=True)
```

```python
stock['month'] = pd.DatetimeIndex(stock['Date']).month
```

```python
nifty.reset_index(inplace=True)
```

```python
nifty['month'] = pd.DatetimeIndex(nifty['Date']).month
```

```python
stock.head()
```

```python
stock_month = stock.groupby('month')
```

```python
stock_month.first()
```

```python
monthly_stock_close_price = list()
for name,group in stock_month:
        monthly_stock_close_price.append(group['Close Price'].sum())
```

```python
monthly_return = pd.DataFrame()
```

```python
monthly_return['Monthly_Stock_close_price'] = monthly_stock_close_price
```

```python
nifty_month = nifty.groupby('month')
```

```python
nifty_month.first()
```

```python
monthly_market_close_price = list()
for name,group in nifty_month:
        monthly_market_close_price.append(group['Close'].sum())
```

```python
monthly_return['Monthly_market_close_price'] = monthly_market_close_price
```

```python
# let us see our monthly close price of stock and market
monthly_return
```

```python
monthly_return['monthly_stock_return'] = monthly_return['Monthly_Stock_close_price'].pct_change()
```

```python
monthly_return['monthly_market_return'] = monthly_return['Monthly_market_close_price'].pct_change()
```

```python
monthly_return
```

```python
monthly_return.drop(0,inplace=True)
```

# lets train our model for monthly stock and market return

```python
x_monthly = monthly_return['monthly_market_return'].values.reshape(-1,1)
y_monthly = monthly_return['monthly_stock_return'].values.reshape(-1,1)
```

```python
lr = LinearRegression()
linear_model_monthly = lr.fit(x_monthly,y_monthly)
```

```python
linear_model_monthly.coef_
```

Therefore Beta Value for monthly return is 1.37515766
